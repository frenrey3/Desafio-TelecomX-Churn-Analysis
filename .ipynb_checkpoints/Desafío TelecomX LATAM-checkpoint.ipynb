#üìå Extracci√≥n

```python
# Extracci√≥n de datos desde URL remota con manejo de errores HTTP y de decodificaci√≥n JSON

import requests
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


resultado = None # Inicializar

try:
    TelecomX_Data = requests.get('https://storage.googleapis.com/enrique-data-bucket/TelecomX_Data.json')
    TelecomX_Data.raise_for_status() # Verificar errores HTTP
    resultado = json.loads(TelecomX_Data.text) # Esto debe ser una lista de diccionarios
    print(f"‚úÖ JSON crudo cargado y parseado. N√∫mero de registros: {len(resultado) if isinstance(resultado, list) else 'No es una lista'}")
except requests.exceptions.RequestException as e_req:
    print(f"‚ùå Error al descargar los datos: {e_req}")
except json.JSONDecodeError as e_json:
    print(f"‚ùå Error al parsear el JSON. El contenido no es un JSON v√°lido: {e_json}")
except Exception as e_other:
    print(f"‚ùå Ocurri√≥ un error inesperado: {e_other}")
```

```python
# Normalizaci√≥n y Creaci√≥n del DataFrame (a√±adiendo una verificaci√≥n)

if resultado is not None and isinstance(resultado, list) and len(resultado) > 0:
    df = pd.json_normalize(resultado)
    print(f"‚úÖ DataFrame normalizado. Dimensiones: {df.shape}")
    df_original_normalized = df.copy()
else:
    print("‚ùå No se pudo crear el DataFrame porque 'resultado' no es una lista v√°lida de datos o est√° vac√≠a.")
    df = pd.DataFrame()

df
```

```python
columns_to_clean = [
    "internet.OnlineSecurity", "internet.OnlineBackup", "internet.DeviceProtection",
    "internet.TechSupport", "internet.StreamingTV", "internet.StreamingMovies", 
    "phone.MultipleLines"
]

# Solo limpiar columnas que existen
existing_columns = [col for col in columns_to_clean if col in df.columns]

df[existing_columns] = df[existing_columns].replace({
    "no internet service": "no",
    "no phone service": "no"
})
```

```python
# Columnas en las que puede aparecer "no internet service" o "no phone service"
columns_to_clean = [
    "onlinesecurity", "onlinebackup", "deviceprotection",
    "techsupport", "streamingtv", "streamingmovies", "multiplelines"
]

# Verificar que las columnas existen antes de limpiar
existing_columns = [col for col in columns_to_clean if col in df.columns]

# Reemplazar valores como "no internet service" y "no phone service" por "no"
df[existing_columns] = df[existing_columns].replace({
    "no internet service": "no",
    "no phone service": "no"
})
```

```python
# Quitar prefijos de columnas
df.columns = df.columns.str.replace(r'^(customer\.|phone\.|internet\.|account\.)', '', regex=True)

# Convertir nombres de columnas a min√∫sculas
df.columns = df.columns.str.lower()

# Mostrar resultado
print("Columnas finales:\n", df.columns.tolist())
```

```python
columns_with_case_issues = [
    "churn", "gender", "partner", "dependents", "phoneservice",
    "multiplelines", "internetservice", "onlinesecurity", "onlinebackup",
    "deviceprotection", "techsupport", "streamingtv", "streamingmovies",
    "contract", "paperlessbilling", "paymentmethod"
]

# Verificar si todas las columnas existen
missing_cols = [col for col in columns_with_case_issues if col not in df.columns]
if missing_cols:
    print("Columnas faltantes:", missing_cols)
else:
    # Convertir valores a min√∫sculas
    df[columns_with_case_issues] = df[columns_with_case_issues].apply(lambda col: col.str.lower())

    # Mostrar valores √∫nicos para verificaci√≥n
    for col in columns_with_case_issues:
        print(f"\nValores √∫nicos en {col}:\n", df[col].unique())
```

```python
# Reemplazar espacios y par√©ntesis en valores categ√≥ricos clave (etiquetas legibles para ML)

columns_to_sanitize = [
    "paymentmethod", "contract", "internetservice"
]

# Aplicar limpieza: reemplazar espacios por guiones bajos y eliminar par√©ntesis
for col in columns_to_sanitize:
    if col in df.columns:
        df[col] = df[col].str.replace(' ', '_')\
                         .str.replace('(', '', regex=False)\
                         .str.replace(')', '', regex=False)

# Verificaci√≥n final de valores √∫nicos por columna clave
columns_final_check = [
    "churn", "gender", "partner", "dependents", "phoneservice",
    "multiplelines", "internetservice", "onlinesecurity", "onlinebackup",
    "deviceprotection", "techsupport", "streamingtv", "streamingmovies",
    "contract", "paperlessbilling", "paymentmethod"
]

for col in columns_final_check:
    if col in df.columns:
        print(f"\nValores √∫nicos en {col}:\n", df[col].unique())
```

```python
# Definir columnas binarias expl√≠citamente
binary_cols = [
    'churn', 'partner', 'dependents', 'phoneservice',
    'multiplelines', 'onlinesecurity', 'onlinebackup',
    'deviceprotection', 'techsupport', 'streamingtv',
    'streamingmovies', 'paperlessbilling'
]

# Conversi√≥n binaria expl√≠cita
df[binary_cols] = df[binary_cols].replace({'yes': 1, 'no': 0})
df[binary_cols] = df[binary_cols].astype(int)

# Conversi√≥n de g√©nero
df[binary_cols] = df[binary_cols].replace({'yes': 1, 'no': 0}).astype(int)
df['gender'] = df['gender'].replace({'female': 0, 'male': 1}).astype(int)

# Convertir columnas categ√≥ricas
category_cols = [
    'multiplelines', 'internetservice',
    'contract', 'paymentmethod'
]
df[category_cols] = df[category_cols].astype('category')

# Asegurar que customerid sea string
df['customerid'] = df['customerid'].astype('string')
```

### 2.1. An√°lisis Descriptivo de los Datos

Para comenzar, realizaremos un an√°lisis descriptivo de los datos, calculando m√©tricas como media, mediana, desviaci√≥n est√°ndar y otras medidas que ayuden a comprender mejor la distribuci√≥n y el comportamiento de los clientes. Se utilizar√° `DataFrame.describe()` y otras funciones de Pandas.

```python
# CELDA: AN√ÅLISIS DESCRIPTIVO GENERAL (C√ÅLCULO DE M√âTRICAS)

if not df.empty:
    print("--- ESTAD√çSTICAS DESCRIPTIVAS DEL DATAFRAME COMPLETO ---")

    print("\n--- Informaci√≥n General (revisi√≥n post-limpieza) ---")
    df.info()

    print("\n--- Estad√≠sticas Descriptivas - Variables Num√©ricas y las convertidas a 0/1 ---")
    try:
        descriptivas_numericas = df.describe(include=[np.number, 'Int64']) 
        display(descriptivas_numericas.transpose())
    except Exception as e:
        print(f"Error al generar describe para num√©ricas/booleanas: {e}")
        try:
            print("Intentando describe() seleccionando dtypes expl√≠citamente...")
            display(df.select_dtypes(include=['float64', 'int64', 'Int64']).describe().transpose())
        except Exception as e_fallback:
            print(f"Describe() con dtypes expl√≠citos tambi√©n fall√≥: {e_fallback}")

    print("\n--- Estad√≠sticas Descriptivas - Variables Categ√≥ricas (object y category) ---")
    try:
        cols_categoricas_reales = df.select_dtypes(include=['object', 'category', 'string']).columns
        if not cols_categoricas_reales.empty:
            cols_para_describir_cat = [col for col in cols_categoricas_reales if col.lower() != 'customerid']
            if cols_para_describir_cat:
                descriptivas_categoricas = df[cols_para_describir_cat].describe().transpose()
                display(descriptivas_categoricas)
            else:
                print("No hay columnas categ√≥ricas (excluyendo customerid) para describir.")
        else:
            print("No hay columnas de tipo 'object', 'category' o 'string' para describir.")
    except Exception as e:
        print(f"Error al generar describe para categ√≥ricas: {e}")
        
    print("\n--- M√©tricas Espec√≠ficas (Ejemplos) ---")
    cols_para_metricas = ['tenure', 'charges.monthly', 'charges.total', 'daily_charges']
    cols_para_metricas_existentes = [col for col in cols_para_metricas if col in df.columns and pd.api.types.is_numeric_dtype(df[col])]
    
    if cols_para_metricas_existentes:
        print("\nMedias:")
        display(df[cols_para_metricas_existentes].mean().to_frame(name='Media'))
        print("\nMedianas:")
        display(df[cols_para_metricas_existentes].median().to_frame(name='Mediana'))
        print("\nDesviaci√≥n Est√°ndar:")
        display(df[cols_para_metricas_existentes].std().to_frame(name='Desv. Est√°ndar'))
    else:
        print("No hay columnas num√©ricas suficientes para calcular m√©tricas espec√≠ficas.")

    print("\n--- Conteo de Valores √önicos por Columna (Cardinalidad) ---")
    try:
        display(df.nunique(dropna=False).sort_values(ascending=False).to_frame(name='Valores √önicos'))
    except Exception as e:
        print(f"Error al calcular nunique: {e}")
else:
    print("El DataFrame est√° vac√≠o. No se puede realizar el an√°lisis descriptivo.")
```

### 2.2. Visualizaci√≥n Descriptiva Univariada

#### 2.2.1. Distribuci√≥n de la Variable Objetivo (Churn)
En este paso, el objetivo es comprender c√≥mo est√° distribuida la variable "churn" (evasi√≥n) entre los clientes. Utilizaremos un gr√°fico de barras para visualizar la proporci√≥n de clientes que permanecieron (0) y los que se dieron de baja (1).

```python
# CELDA: VISUALIZACI√ìN MEJORADA DE LA DISTRIBUCI√ìN DE CHURN (CON GRID FINA)
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker # Necesario para los locators

col_churn = 'churn' 

if not df.empty and col_churn in df.columns:
    plt.figure(figsize=(8, 6))
    
    churn_labels_for_plot = df[col_churn].map({0: 'Permanecen (0)', 1: 'Evaden (1)'})
    
    ax = sns.countplot(
        x=churn_labels_for_plot,
        hue=churn_labels_for_plot, 
        palette=['#3498db', '#e74c3c'],
        order=['Permanecen (0)', 'Evaden (1)'],
        legend=False
    )
    
    plt.title('Distribuci√≥n de Evasi√≥n de Clientes (Churn)', 
              fontsize=18, fontweight='bold', pad=25) # Aument√© un poco el pad y fontsize
    plt.xlabel('Estado del Cliente', fontsize=14, fontweight='bold', labelpad=15) # labelpad para espacio
    plt.ylabel('N√∫mero de Clientes', fontsize=14, fontweight='bold', labelpad=15) # labelpad para espacio
    
    total = len(df[col_churn])
    max_height = 0 # Para calcular el l√≠mite superior del eje Y
    
    for p in ax.patches:
        count = int(p.get_height())
        if count > max_height: # Actualizar max_height
            max_height = count
        percentage = f'{100 * count / total:.1f}%'
        
        ax.annotate(f'{count}\n({percentage})', 
                    (p.get_x() + p.get_width() / 2., count), 
                    ha='center',
                    va='bottom', 
                    fontsize=10, # Ligeramente m√°s peque√±o para no saturar
                    fontweight='bold',
                    color='#333333', # Un gris oscuro para el texto
                    xytext=(0, 5),  # 5 puntos arriba de la barra
                    textcoords='offset points')
    
    ax.set_ylim(0, max_height * 1.15) # 15% m√°s espacio arriba para el texto

    # --- ESTILO DE SPINES Y TICKS ---
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    # ax.spines['left'].set_visible(False) # Considerar mantenerlo para referencia del eje Y si no poner grid fina
    
    ax.tick_params(axis='x', 
                   length=0, 
                   labelsize=12, # Ligeramente m√°s grande
                   labelcolor='#2c3e50',
                   pad=10) 
    
    ax.tick_params(axis='y', 
                   labelsize=11, # Ligeramente m√°s grande
                   labelcolor='#2c3e50',
                   left=True) # Mantener los ticks si quieres grid fina

    ax.spines['bottom'].set_linewidth(1.5)
    ax.spines['bottom'].set_color('#34495e')
    ax.spines['left'].set_linewidth(1.0) # Si decides mantenerlo
    ax.spines['left'].set_color('#34495e') # Si decides mantenerlo

    # --- CONFIGURACI√ìN DE LA GRID FINA ---
    # Grid principal (l√≠neas m√°s gruesas o visibles)
    ax.yaxis.grid(True, linestyle='-', linewidth=0.7, alpha=0.6, color='#cccccc') # Grid principal m√°s visible
    
    # Grid secundaria/fina (l√≠neas m√°s sutiles entre las principales)
    # Para esto, a menudo necesitas definir los locators.
    # Por ejemplo, si quieres una l√≠nea cada 500 clientes y una m√°s fina cada 100:
    ax.yaxis.set_major_locator(ticker.MultipleLocator(base=1000)) # L√≠nea principal cada 1000
    ax.yaxis.set_minor_locator(ticker.MultipleLocator(base=250))  # L√≠nea fina cada 250

    # Mostrar la grid menor (fina)
    ax.grid(which='minor', axis='y', linestyle=':', linewidth=0.5, alpha=0.5, color='#dddddd')
    
    # Asegurar que la grid est√© detr√°s de las barras
    ax.set_axisbelow(True)
    
    # Quitar la l√≠nea duplicada de plt.grid(True) que ten√≠a al final
    # La configuraci√≥n de ax.yaxis.grid() y ax.grid(which='minor'...) es m√°s espec√≠fica.
    
    plt.tight_layout() # Ajustar para que todo quepa bien
    plt.show()
    
else:
    print("DataFrame vac√≠o o la columna 'churn' no existe.")
```

### 2.2.2. Distribuci√≥n de Evasi√≥n (Churn) por Variables Categ√≥ricas

Exploraremos c√≥mo se distribuye la evasi√≥n (`churn`) seg√∫n diferentes caracter√≠sticas categ√≥ricas de los clientes. Esto nos ayudar√° a identificar si ciertos grupos de clientes tienen una mayor o menor propensi√≥n a la evasi√≥n.

**A. Visualizaci√≥n de Conteos Absolutos de Churn por Categor√≠a** 

```python
# CELDA: VISUALIZACI√ìN DE CHURN POR VARIABLES CATEG√ìRICAS (COUNTPLOTS MEJORADOS)

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd # Aseg√∫rate que pd est√© definido (DataFrame)
import numpy as np  # Aseg√∫rate que np est√© definido (para isinstance con np.integer)
from IPython.display import display

col_churn = 'churn' # Tu columna de evasi√≥n (0 o 1)

if not df.empty and col_churn in df.columns:
    print("--- AN√ÅLISIS DE EVASI√ìN POR VARIABLES CATEG√ìRICAS (CONTEOS) ---")
    
    categorical_multivalue_cols = df.select_dtypes(include=['object', 'category', 'string']).columns.tolist()
    if 'customerid' in categorical_multivalue_cols:
        categorical_multivalue_cols.remove('customerid')

    binary_cols_as_cat_for_plot = [
        'gender', 'seniorcitizen', 'partner', 'dependents', 
        'phoneservice', 'multiplelines', 'onlinesecurity', 
        'onlinebackup', 'deviceprotection', 'techsupport', 
        'streamingtv', 'streamingmovies', 'paperlessbilling'
    ]
    binary_cols_as_cat_for_plot = [col for col in binary_cols_as_cat_for_plot if col in df.columns]
    
    features_to_analyze = sorted(list(set(categorical_multivalue_cols + binary_cols_as_cat_for_plot)))

    churn_legend_map = {0: 'Permanecen', 1: 'Evaden'} # Para la leyenda

    for feature in features_to_analyze:
        if feature in df.columns and feature != col_churn:
            
            # --- Preparar datos para el plot ---
            data_plot_cat = df.copy()
            order_for_plot = None
            feature_label = feature.capitalize().replace("_", " ") # Para t√≠tulos y etiquetas de eje

            # Mapear valores num√©ricos binarios a etiquetas m√°s descriptivas para el eje del gr√°fico
            if data_plot_cat[feature].dtype in ['int64', 'Int64'] and df[feature].nunique() == 2:
                unique_vals_feat = sorted(df[feature].dropna().unique().tolist())
                if len(unique_vals_feat) == 2 and all(isinstance(x, (int, np.integer)) for x in unique_vals_feat):
                    map_dict_feat = {unique_vals_feat[0]: f'No ({unique_vals_feat[0]})', 
                                     unique_vals_feat[1]: f'S√≠ ({unique_vals_feat[1]})'}
                    if feature == 'gender': map_dict_feat = {0: 'Female', 1: 'Male'} # Ejemplo espec√≠fico
                    elif feature == 'seniorcitizen': map_dict_feat = {0: 'No Senior', 1: 'Senior'}
                    
                    data_plot_cat[feature] = data_plot_cat[feature].map(map_dict_feat)
                    order_for_plot = [map_dict_feat[val] for val in unique_vals_feat if val in map_dict_feat]
            
            # Si no se defini√≥ un orden espec√≠fico, ordenar por frecuencia
            if order_for_plot is None:
                order_for_plot = data_plot_cat[feature].value_counts().index

            # --- Crear el Countplot ---
            plt.figure(figsize=(10, max(5, len(order_for_plot) * 0.6) )) # Ajustar altura din√°micamente
            
            ax = sns.countplot(
                data=data_plot_cat, 
                y=feature, 
                hue=df[col_churn].map(churn_legend_map), # Usar churn original mapeado para hue
                palette={'Permanecen': '#3498db', 'Evaden': '#e74c3c'}, # Paleta por nombre de categor√≠a
                order=order_for_plot
            )
            
            plt.title(f'Distribuci√≥n de Churn por {feature_label}', fontsize=16, fontweight='bold', pad=20)
            plt.xlabel('N√∫mero de Clientes', fontsize=13, fontweight='bold', labelpad=10)
            plt.ylabel(feature_label, fontsize=13, fontweight='bold', labelpad=10)
            
            ax.tick_params(axis='both', labelsize=11)
            sns.despine() # Quita spines superior y derecho por defecto

            # Mejorar leyenda
            handles, labels = ax.get_legend_handles_labels()
            ax.legend(handles, labels, title='Estado de Churn', title_fontsize='12', 
                      loc='center left', bbox_to_anchor=(1, 0.5)) # Leyenda fuera del gr√°fico

            # Anotaciones (opcional para countplot con hue, puede ser denso)
            # Si decides anotarlo:
            # for p in ax.patches:
            #     width = p.get_width()
            #     if width > 0: # Solo anotar si hay valor
            #         ax.text(width + (ax.get_xlim()[1] * 0.01), # Un peque√±o offset desde el final de la barra
            #                 p.get_y() + p.get_height() / 2.,
            #                 f'{int(width)}',
            #                 ha='left', va='center', fontsize=9)
            
            plt.tight_layout(rect=[0, 0, 0.85, 1]) # Ajustar para que quepa la leyenda
            plt.show()
            print("-" * 70)
else:
    print(f"DataFrame vac√≠o o la columna '{col_churn}' no disponible.")
```

**B. An√°lisis de Tasas de Evasi√≥n por Categor√≠a**
Para profundizar, calcularemos y visualizaremos la tasa de evasi√≥n espec√≠fica para cada categor√≠a de las variables.

```python
# CELDA: TASAS DE EVASI√ìN Y GR√ÅFICOS APILADOS POR VARIABLES CATEG√ìRICAS

col_churn = 'churn'

if not df.empty and col_churn in df.columns:
    print("\n--- AN√ÅLISIS DE TASA DE EVASI√ìN POR VARIABLES CATEG√ìRICAS ---")

    # Reusar features_to_analyze de la celda anterior o redefinirla aqu√≠
    # (Aseg√∫rate de que 'features_to_analyze' est√© definida y sea la misma lista)
    if 'features_to_analyze' not in locals(): # Si esta celda se ejecuta independientemente
        categorical_multivalue_cols = df.select_dtypes(include=['object', 'category', 'string']).columns.tolist()
        if 'customerid' in categorical_multivalue_cols: categorical_multivalue_cols.remove('customerid')
        binary_cols_as_cat_for_plot = [
            'gender', 'seniorcitizen', 'partner', 'dependents', 'phoneservice', 'multiplelines', 
            'onlinesecurity', 'onlinebackup', 'deviceprotection', 'techsupport', 
            'streamingtv', 'streamingmovies', 'paperlessbilling'
        ]
        binary_cols_as_cat_for_plot = [col for col in binary_cols_as_cat_for_plot if col in df.columns]
        features_to_analyze = sorted(list(set(categorical_multivalue_cols + binary_cols_as_cat_for_plot)))


    for feature in features_to_analyze:
        if feature in df.columns and feature != col_churn:
            feature_label = feature.capitalize().replace("_", " ")

            if df[feature].isnull().sum() == 0 and df[col_churn].isnull().sum() == 0:
                print(f"\n--- An√°lisis para: {feature_label} ---")
                
                # Tabla Cruzada (ya la ten√≠as y es √∫til)
                crosstab_df = pd.crosstab(df[feature], df[col_churn])
                crosstab_df.rename(columns={0: 'Permanecen', 1: 'Evaden'}, inplace=True)
                print("Tabla Cruzada (Conteos):")
                display(crosstab_df)

                # Porcentajes (normalize='index')
                crosstab_norm = pd.crosstab(df[feature], df[col_churn], normalize='index') * 100
                crosstab_norm.rename(columns={0:'% Permanecen', 1:'% Evaden'}, inplace=True)
                print("\nPorcentaje de Evasi√≥n/Permanencia por categor√≠a:")
                display(crosstab_norm.sort_values(by='% Evaden', ascending=False))
                
                # --- Gr√°fico de barras de la Tasa de Evasi√≥n ---
                plt.figure(figsize=(8, max(4, crosstab_norm.shape[0]*0.4))) # Ajustar altura
                # Mapear √≠ndices num√©ricos (0/1) a etiquetas para el eje Y del gr√°fico de tasas
                plot_data_rate = crosstab_norm['% Evaden'].sort_values(ascending=True)
                y_labels_rate = plot_data_rate.index
                if df[feature].dtype in ['int64', 'Int64'] and df[feature].nunique() == 2:
                     unique_vals_feat = sorted(df[feature].dropna().unique().tolist())
                     map_dict_feat_rate = {unique_vals_feat[0]: f'No ({unique_vals_feat[0]})', 
                                           unique_vals_feat[1]: f'S√≠ ({unique_vals_feat[1]})'}
                     if feature == 'gender': map_dict_feat_rate = {0: 'Female', 1: 'Male'}
                     elif feature == 'seniorcitizen': map_dict_feat_rate = {0: 'No Senior', 1: 'Senior'}
                     y_labels_rate = plot_data_rate.index.map(map_dict_feat_rate)


                ax_rate = plot_data_rate.plot(kind='barh', color='#e74c3c', width=0.7)
                plt.title(f'Tasa de Evasi√≥n (%) por {feature_label}', fontsize=15, fontweight='bold', pad=15)
                plt.xlabel('Tasa de Evasi√≥n (%)', fontsize=12, fontweight='bold')
                plt.ylabel(feature_label, fontsize=12, fontweight='bold')
                ax_rate.set_yticklabels(y_labels_rate) # Aplicar etiquetas mapeadas
                plt.xlim(0, max(plot_data_rate.max() * 1.2, 20) if not plot_data_rate.empty else 100) # Ajustar l√≠mite X

                for index, value in enumerate(plot_data_rate):
                    plt.text(value + 0.5 , index, f'{value:.1f}%', va='center', ha='left', fontsize=9)
                
                sns.despine()
                plt.tight_layout()
                plt.show()

                # --- OPCIONAL: Gr√°fico de Barras Apiladas al 100% ---
                print(f"\n(Opcional) Gr√°fico de Barras Apiladas al 100% para {feature_label}:")
                ax_stacked = crosstab_norm[['% Permanecen', '% Evaden']].sort_values(by='% Evaden').plot(
                    kind='barh', 
                    stacked=True, 
                    figsize=(10, max(4, crosstab_norm.shape[0]*0.4)), 
                    color={'% Permanecen': '#3498db', '% Evaden': '#e74c3c'},
                    width=0.8
                )
                plt.title(f'Proporci√≥n de Churn por {feature_label}', fontsize=15, fontweight='bold', pad=20)
                plt.xlabel('Porcentaje de Clientes (%)', fontsize=12, fontweight='bold')
                plt.ylabel(feature_label, fontsize=12, fontweight='bold')
                ax_stacked.set_yticklabels(y_labels_rate) # Aplicar etiquetas mapeadas si es necesario
                plt.legend(title='Estado de Churn', bbox_to_anchor=(1.02, 1), loc='upper left')
                plt.xlim(0, 100)
                
                # Anotaciones para barras apiladas (pueden ser un poco complejas de ajustar)
                for n, (idx, row) in enumerate(crosstab_norm[['% Permanecen', '% Evaden']].sort_values(by='% Evaden').iterrows()):
                    cumulative_width = 0
                    for i, (colname, value) in enumerate(row.items()):
                        if value > 5: # Solo anotar si el segmento es mayor al 5% para no saturar
                            bar_width = value
                            label = f'{value:.0f}%' # Porcentaje redondeado
                            x_pos = cumulative_width + bar_width / 2
                            ax_stacked.text(x_pos, n, label, ha='center', va='center', color='white', fontsize=8, fontweight='bold')
                        cumulative_width += value
                
                sns.despine()
                plt.tight_layout(rect=[0, 0, 0.85, 1])
                plt.show()
                print("-" * 70)

            else:
                print(f"Advertencia: La columna '{feature}' o '{col_churn}' contiene NaNs, omitiendo crosstab y gr√°ficos de tasas.")
        else:
            print(f"Advertencia: La columna categ√≥rica '{feature}' no existe en el DataFrame.")
else:
    print(f"DataFrame vac√≠o o la columna '{col_churn}' no disponible.")
```
